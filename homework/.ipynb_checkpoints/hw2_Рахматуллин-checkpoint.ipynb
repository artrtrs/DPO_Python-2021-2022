### Задача 4

В математике существует так называемая последовательность чисел Фибоначчи. Выглядит она так: 1, 1, 2, 3, 5, 8, 13, ...

Каждое последующее число равно сумме двух предыдущих, а первые два числа Фибоначчи - две единицы.
Запросите с клавиатуры число N и запишите в список первые N чисел Фибоначчи.

```python
fibN = int(input('Сколько чисел Фибоначчи нужно сгенерировать? '))

# объявляем список
fib = [None]*fibN
fib[0] = 1
fib[1] = 1

for i in range(2,fibN):
    fib[i] = fib[i-2]+fib[i-1]

print(fib)
```

### Задача 5 Подсчет слов
Давайте теперь поработаем с настоящим файлом и действительно посчитаем в нем слова. Мы загрузим метаданные почтового сервера университета Мичигана. И попробуем найти, с какого адреса ушло больше всего писем.

```python
# импортируем библиотеку для доступа к файлам в интернете
import requests
import re
# в переменной mbox хранится текст для работы
mbox = requests.get('http://www.py4inf.com/code/mbox.txt').text

# находим почтовые адреса
mail_lst = re.findall(r'From (.+@\S+)', mbox) 

# создаем множество из почтовых адресов, чтобы в дальнейшем использовать его в цикле для подсчета 
# отправленных писем с этого адреса
unique_mails = set(mail_lst)

# создаем словарь для хранения в нем пар e-mail: количество отправленных писем
mails = dict()

# пишем цикл для подсчета количества отправленных писем
for i in unique_mails:
    mails[i] = 0
    for j in mail_lst:
        if j == i:
            mails[i] += 1

maxmails = max(mails, key=mails.get)
print(maxmails) 
```

### Задача 6 Посчитать среднюю метрику X-DSPAM-Confidence для письма. 
### (Кажется, это что-то похожее на вероятность того, что письмо - не спам)

Давайте воспользуемся предыдущим кодом и достанем те строчки, где эта метрика записана:

```python
# импортируем библиотеку для доступа к файлам в интернете
import requests
# в переменной mbox хранится текст для работы
mbox = requests.get('http://www.py4inf.com/code/mbox.txt').text

import re

dspam = re.findall(r'X-DSPAM-Confidence: (0.\d+)', mbox)

len_dspam = len(dspam)
sum_dspam = 0

# считаем сумму всех значений метрики
for x in dspam:
    sum_dspam += float(x)

# считаем среднее значение метрики
av_dspam = sum_dspam / len_dspam

print(av_dspam)



```

### Задача 7 Объединение словарей

Напишите программу, которая объединяет значения из двух списков.

Ввод:

    shops = [{'товар': 'яблоки', 'количество': 400}, {'товар': 'конфеты', 'количество': 300}, {'товар': 'яблоки', 'количество': 750}]


Вывод:

    {'яблоки': 1150, 'конфеты': 300}

```python
shops = [{'товар': 'яблоки', 'количество': 400}, 
         {'товар': 'конфеты', 'количество': 300}, 
         {'товар': 'яблоки', 'количество': 750}]

new_shops = {}

len_shops = len(shops)

i = 0

while i < len_shops:
    # сначала в новом словаре задаем структуру вида 'яблоки': 0 
    # инициализируем нулем количество для каждого товара для того, чтобы использовать в следующем цикле для суммирования
    new_shops[shops[i]['товар']] = 0
    i += 1
    
j = 0
while j < len_shops:
    # просуммируем количество для каждого товара
    new_shops[shops[j]['товар']] += shops[j]['количество']
    j += 1

    
print(new_shops)
    
```

### Задача 8 Найти все числа от 1000 до 3000 включительно, все цифры которых четные.

Программа должна выдавать результат в виде разделенной запятыми строки

```python
lst = [x for x in range(1000,3001)]

#print(lst)

lst_even = []

i = 0

for i in lst:
    # проверяем на четность каждый разряд числа
    if i % 2 == 0 and int(i / 1000) % 2 == 0 and int(i / 100) % 2 == 0 and int(i / 10) % 2 == 0:
        lst_even.append(i)
        #print(i)

print(*lst_even, sep=',')
```

### Задача 11 Дан массив из чисел. Отсортируйте его, не используя встроенных функций. (циклами for, например). Ожидаю получить сортировку пузырьком)

Вход: На вход дана последовательность из N случайных чисел. Например [48, 59, 9, 90, 15, 58, 19, 49, 13, 7]


Выход:

    Отсортированная последовательность [7, 9, 13, 15, 19, 48, 49, 58, 59, 90]

```python
from random import randint
 
N = 10
a = []
for i in range(N):
    a.append(randint(1, 99))
#print(a)

len_a = len(a)

i = 0
j = 0

# пузырьковая сортировка
while i < len_a:
    for j in range(len_a-1):
        if a[j] > a[j+1]:
            a[j], a[j+1] = a[j+1], a[j]
            #print(a)
        j += 1
    i += 1
    
print(a)
```

### Задача 12 Про Поросёнка Петра

На плоскости в точке (0,0) стоит Поросёнок Пётр. Он умеет ходить налево, направо, вверх и вниз. Расстояние его прохода в какую-либо сторону измеряется в шагах. Когда он идет вправо, его первая координата увеличивается, когда влево - уменьшается. Когда он идет вверх, его вторая координата увеличивается, а когда вниз - уменьшается. 

С клавиатуры считывается число N - число ходов, которые сделает Пётр. После чего на каждом шаге спрашивается, сколько шагов и в какую сторону за этот ход Пётр сделает. Так происходит, пока Пётр не осуществит все N ходов.

Программа должна вывести, сколько шагов Пётр должен был бы сделать, чтобы кратчайшим путем прибыть из свое начальной точки (0,0) в свою конечную точку. Напоминание: Пётр умеет ходить только вверх-вниз, и влево-вправо, но не по диагонали.

Пример ввода: 

    Введите N: 3
    Ход 1: Вверх 1
    Ход 2: Вниз 1
    Ход 3: Вверх 1
    
Пример вывода:


Пётр находится на расстоянии 1 от (0,0)

```python
import re

N = int(input('Введите N: '))

# заводим и инициализируем нулями словарь для расчета координат после каждого шага
coords = {}
coords['x'] = 0
coords['y'] = 0

# счетчик ошибок на случай, если ходы введены не по шаблону
nErrors = 0

for _ in range(N):
    str1 = input('Ход '+str(_+1)+': ')
    # проверяем введенную строку на соответствие шаблону
    if re.match(r'Вверх\s\d|Вниз\s\d|Вправо\s\d|Влево\s\d', str1) is not None:
        # если строка соответствует шаблону, то с помощью словаря подсчитываем координаты  x,y
        str1 = str1.split()
        if str1[0] == 'Вверх':
            coords['y'] += int(str1[1])
        elif str1[0] == 'Вниз':
            coords['y'] -= int(str1[1])
        elif str1[0] == 'Вправо':
            coords['x'] += int(str1[1])
        elif str1[0] == 'Влево':
            coords['x'] -= int(str1[1])
    else:
        nErrors = 1
        print('Ваш ввод координат не соответствует шаблону. \nНеобходимо вводить координаты в виде: \nВверх 1\nВниз 1\nВправо 2')
        break

if nErrors == 0:
        distance = abs(coords['x']) + abs(coords['y'])
        print('Петр находится на расстоянии '+str(distance)+' от (0,0)')
```
